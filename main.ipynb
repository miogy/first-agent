```python
# openai 에러 > 커널 선택후 설치 확인

# import sys
# print("python:", sys.executable)

# import openai
# print("openai version:", openai.__version__)
# print("has OpenAI class:", hasattr(openai, "OpenAI"))

import openai

client = openai.OpenAI()

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role":"user",
            "content":"What is the capital of Greece"
        }
    ]
) # ChatCompletion > choices > messages > ChatCompletionMessage > content 확인(LLM이 보낸 응답)

message = response.choices[0].message.content

message


```
